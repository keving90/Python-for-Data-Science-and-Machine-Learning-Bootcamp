# Pandas Built-in Data Visualization

#### Pandas allows you to create data visualization directly off a dataframe.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# Creating a histogram in Pandas is simple
df1['A'].hist(bins=30)
```

```python
# The histogram above doesn't look very nice
# Import Seaborn at implicitly get better looking visualizations
# It's not working. GG
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist')
```

```python
# Another way to call hist()
# We will use this method
df1['A'].plot.hist()
```

```python
df2.head()
```

```python
# Use alpha to make it more transparent
df2.plot.area(alpha=0.4)
```

```python
# Uses the index as a category
df2.plot.bar()
```

```python
df2
```

```python
# Can stack the data
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
# Line plot
# Can use matplotlib features to alter the plot
df1.plot.line(x=df1.index, y='B', figsize=(12, 3), lw=1)
```

```python
# Can alter the color based on the values in a column
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# Can do the same as above, but with size instead of color
# The points will be too small, so you will need to add a multiplying factor
df1.plot.scatter(x='A', y='B', s=df1['C']*50)
```

```python
# Box plots
df2.plot.box()
```

```python
# Create a new dataframe to use with our hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df.head()
```

```python
# Hexagonal bin plot for bivariate data
df.plot.hexbin(x='a', y='b')
```

```python
# Use gridsize to make the bins larger
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
# Kernel Density Estimation plot
df2['a'].plot.kde()
```

```python
# Same as kde()
df2['a'].plot.density()
```

```python
# Include all columns
df2.plot.density()
```

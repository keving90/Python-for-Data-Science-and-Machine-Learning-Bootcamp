```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

### There are two ways of creating plots - Functional Method, and Object-Oriented Method

### Functional Method

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show() - This must be at the end of your code to show the plot if you're not in Jupyter
```

```python
# Multiplots on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

### Object-Oriented Method

```python
# Create figure objects and call methods from them
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

# List corresponds to:

# ['how far right the bottom left corner is',
#  'how high the bottom left corner is',
#  'how wide the graph is',
#  'how tall the graph is']

# Values must be between 0 and 1
# Negative values are OK!

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Matplotlib Part 2

```python
# Creating subplots
# With subplots(), you can specify the number of rows and columns
# Take advantage of tuple unpacking 
fig, axes = plt.subplots(nrows=1,ncols=2)

# axes.plot(x,y)

# Can iterate over the axes object to quickly plot over each axis
# for current_ax in axes:
#     current_ax.plot(x,y)

# Can reference by index
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.subplots(nrows=3, ncols=3)

# Use tight_layout() to create space in a crowded area
# Make sure the method is from the matplotlib library, not the plot pbject (plt instead of fig)
plt.tight_layout()
```

### Figure Size and DPI

```python
# The tuple is (width, height) in inches
# DPI is Dots (Pixels) per Square Inch

fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
# You can save a figure to a file
# Just pass in the extension
# Can specify DPI
fig.savefig('my_picture.png', dpi=200)
```

```python
# Can plot two lines on the same graph

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')

# Add a legend
# Need to add legend labels in the plot() methods with 'label='
# Can specify legend location on the plot in the method legend()
ax.legend(loc=0)

# Specify legend location with tuple (x,y)
# (x,y) refers to the coordinates of the lower left corner of the legend
ax.legend(loc=(0.1,0.1))
```

# Matplotlib Part 3

```python
# Setting colors
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Basic color descriptions will work
# ax.plot(x,y,color='green')

# Can also use RGB hex codes (begin with # sign)
# Google 'RGB hex codes' for a list of codes
# ax.plot(x,y,color='#FF8C00')

# Can specify line width (default is 1)
# alpha argument allows you to control how transparent the line is
ax.plot(x,y,color='#FF8C00', linewidth=3, alpha=0.3)

# Can type in 'lw' for 'linewidth'
# ax.plot(x,y,color='#FF8C00', lw=3, alpha=0.1)
```

```python
# Line styles
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3, alpha=0.5, linestyle=':')

# Can use 'ls' instead of 'linestyle'
# ax.plot(x,y,color='purple', lw=3, alpha=0.5, ls=':')
```

```python
# Markers
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Can make many specifications to the markers
# See the instructor's notebook for more variations
ax.plot(x,y,color='blue', lw=1, alpha=0.5, ls='-', marker='o', markersize='20',
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

### Control over axis appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue', lw=2, alpha=0.5, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
